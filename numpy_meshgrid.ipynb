#### Purpose of Meshgrid
- Meshgrid is very usefull to evaluate on a grid.We can apply fun to the points of a meshgrid to plot a function

```python
x = np.linspace(-40,40,100)
y = np.linspace(-50,50,90)
XX,YY = np.meshgrid(x,y)
print(XX.shape,YY.shape)
```

```python
z = (XX**2) + (YY**2)

plt.contourf(XX,YY,z)
```

```python
z = np.random.random((90,100))

plt.contourf(XX,YY,z)
```
